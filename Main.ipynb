```python
import jaya
```

```python
def select_function(inc,f):
    if inc==1: 
        function_selected=jaya.jaya(f)
        return function_selected
    if inc==2:
        function_selected=jaya.jaya_modified(f)
        return function_selected
    else:
        return 0
```

```python
for k in range(1,3):
    for i in range(1,24):   
        n = 30   #no of runs
        for j in range(n):
            print('Optimising for F'+str(i)+ ' Run '+str(j))
            select_function(k,i)
```
